

# Face Detection using OpenCV

In this notebook, we explore the fascinating world of computer vision and face detection using OpenCV, a powerful computer vision library in Python. The primary goal is to leverage a pre-trained face classifier to detect faces in an input image. We'll walk through the process of loading the classifier, reading an image, converting it to grayscale, and performing face detection. Visualizations using rectangles will showcase the detected faces.

```python
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained face classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Read the input image
image_path = '.\\Data\\1.jpg'
img = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Perform face detection
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

# Draw rectangles around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the output image with faces detected using matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.axis('off')
plt.show()
```
